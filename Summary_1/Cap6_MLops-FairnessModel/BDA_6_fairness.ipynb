# MLOps: Sistemi di ML in produzione

## Ciclo di vita di un progetto di ML



Machine Learning: 

- Il machine learning è la scienza degli algoritmi informatici che imparano e migliorano automaticamente dall'esperienza anziché essere programmati esplicitamente. 

- Gli algoritmi analizzano dati di esempio (dati di addestramento) per costruire un modello software in grado di fare previsioni. 

- I dati di test sono in qualche modo correlati ai dati precedenti.

### Quando usare ML

Problemi complessi che richiedono molti parametri e dsi deve capire come fargli prendere decisioni:

- il rilevamento dello spam: 
    - impossibile scrivere un codice che rilevi i messaggi di spam e lasci passare i messaggi genuini nella casella di posta.

- Predire se un mutuatario ripagherà un prestito. 



- Quando il problema cambia costantemente, il codice di programmazione deve essere regolarmente aggiornato. Quando si tratta di un problema di percezione.

- Le principali fonti di costo nel machine learning includono:
- la raccolta
- la preparazione
- la pulizia dei dati
- l'addestramento del modello
- la costruzione e l'esecuzione dell'infrastruttura per servire e monitorare il modello
- le risorse umane per la manutenzione.

### Perchè i progetti ML falliscono:

- Mancanza di talento esperto:

    - data science e l'ingegneria del machine learning sono discipline relativamente nuove.
    - Non esiste ancora un modo standard per insegnarle e procedere allo sviluppo

- Mancanza di supporto da parte della leadership:

- ricercatori e gli ingegneri del software lavorano in modi molto diversi.
    
    - i team di ingegneria del software lavorano con obiettivi chiaramente definiti e poca incertezza.

    - ricercatori lavorano in un contesto di elevata incertezza
    - la maggior parte degli esperimenti non produce alcun risultato tangibile
        - percezione della leadership. del lavoro del ricercatore come tempo e risorse sprecate.

- Mancanza di infrastruttura dati:
    - Qualità dati cruciale per successo progetto ML.
    - Infrastruttura dati aziendale deve fornire un modo semplice per ottenere dati addestramento.
    - qualità dati disponibili in produzione.
    

- Sfida etichettatura dati (ground truth):

    - Etichettatura personalizzata per ogni progetto.
    - Investimento tempo per etichettatura e sviluppo strumenti.
    - Dati etichettati di scarsa qualità senza validazione adeguata.
    

## MLOps

Il deploy di un modello di ML in un sistema di Machine Learning in produzione (in grado di generare un valore economico) è un processo complesso che richiede competenze diverse e specifiche.

- IL modello va integrato in un software;

**decadimento delle prestazioni del modello**

- con il passare del tempo il modello va monitorato poiche i dati cambiano a causa dei cambiamenti nella vita reale;

    - il modello va retrainato;
    
**Qualità dei dati**

- I modelli sono basati sui dati e sono sensibili alla semantica, la quantità e la completezza dei dati;

**Località**

- Se il modello è stato allenato su dati di un paese e viene utilizzato in un altro paese, potrebbe non funzionare bene;

**MLOps Definizione**:

- MLOps è un approccio per la creazione, il test e il rilascio di applicazioni di Machine Learning attraverso l'automazione e il monitoraggio in ogni fase del ciclo di vita di Machine Learning.

**Mitigazione del Rischio**

MLOps per mitigazione rischio:

- Introduzione modelli ML senza infrastruttura MLOps rischiosa:

  - Prestazioni valutabili solo in produzione.
  - Dati addestramento riflettono produzione.
  - Cambiamenti in produzione possono ridurre prestazioni
  - Sensibilità prestazioni ML all'ambiente produzione, inclusa versione software.
  - Modelli basati su software open source, necessarie stesse versioni in produzione.
- Introduzione modelli non fine ciclo ML, ma inizio 
    - monitoraggio prestazioni
        - garanzia comportamento atteso del modello
